```python
# Basic Pandas operations

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Pandas Series

print("=== Pandas Series Example ===")

# Creating a Series from a list
ages = pd.Series([25, 30, 35, 40], name="Age")

print("\nSeries created from a list:")
print(ages)

# Accessing elements
print("\nFirst element of Series:", ages[0])
print("Last element of Series:", ages.iloc[-1])

# Performing operations on Series
print("\nAdding 5 to each element in Series:")
print(ages + 5)
```

```python
# Create a dictionary of data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Score": [85, 90, 95, 88]
}

# Create a DataFrame
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
# Selecting a single column
print("\nSelecting 'Name' column:")
print(df["Name"])
```

```python
# Selecting multiple columns
print("\nSelecting 'Name' and 'Score' columns:")
print(df[["Name", "Score"]])
```

```python
# Filtering rows (Age > 30)
print("\nFiltering rows where Age > 30:")
print(df[df["Age"] > 30])
```

```python
# Adding a new column (Grade based on Score)
df["Grade"] = ["B", "A", "A+", "A"]

print("\nDataFrame after adding Grade column:")
print(df)
```

```python
# Basic statistics
print("\nBasic statistics of numeric columns:")
print(df.describe())
```

```python
# Sorting by Score
print("\nSorting DataFrame by 'Score':")
print(df.sort_values(by="Score", ascending=False))
```

```python
# Plotting the Series
ages.plot(kind="line", title="Ages (Series Plot)", marker="o")
plt.xlabel("Index")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```

```python
# Plotting a column (Score)
df["Score"].plot(kind="bar", title="Scores of Students")
plt.xticks(range(len(df)), df["Name"], rotation=0)
plt.xlabel("Students")
plt.ylabel("Score")
plt.show()
```

```python
# Plotting multiple columns (Age vs Score)
df.plot(x="Name", y=["Age", "Score"], kind="line", marker="o", title="Age vs Score")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
