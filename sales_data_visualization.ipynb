```python
#  Sales Data Visualization with Pandas & Matplotlib

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Create Sample Sales Data or Load Sales Data from CSV
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product_A_Sales": [120, 150, 170, 130, 160, 180],
    "Product_B_Sales": [100, 120, 140, 110, 150, 170],
    "Product_C_Sales": [90, 110, 130, 100, 120, 150],
}
df = pd.DataFrame(data)
#df = pd.read_csv("sales_data.csv")
print("ðŸ“Œ Sales Data:")
print(df)
```

```python
# 2. Line Plot â€“ Sales Trend Over Months
plt.figure(figsize=(8,5))
plt.plot(df["Month"], df["Product_A_Sales"], marker="o", label="Product A")
plt.plot(df["Month"], df["Product_B_Sales"], marker="s", label="Product B")
plt.plot(df["Month"], df["Product_C_Sales"], marker="^", label="Product C")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Bar Chart â€“ Compare Products by Month
df.plot(x="Month", kind="bar", figsize=(8,5))
plt.title("Sales Comparison by Month")
plt.ylabel("Sales")
plt.show()
```

```python

# 4. Pie Chart â€“ Total Sales Contribution
total_sales = df[["Product_A_Sales", "Product_B_Sales", "Product_C_Sales"]].sum()
total_sales.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6), startangle=90)
plt.title("Total Sales Contribution by Product")
plt.ylabel("")  # Remove y-label
plt.show()
```
