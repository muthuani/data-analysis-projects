```python
# Iris Dataset Exploratory Data Analysis (EDA)

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
#url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
url = "iris.csv"
df = pd.read_csv(url)
```

```python
# 1. Basic Exploration

print("ðŸ”¹ First 5 rows of the dataset:")
display(df.head())

print("\nðŸ”¹ Dataset Info:")
print(df.info())

print("\nðŸ”¹ Shape of dataset:", df.shape)

print("\nðŸ”¹ Summary Statistics:")
display(df.describe())

print("\nðŸ”¹ Unique Species:", df['species'].unique())
print("\nðŸ”¹ Count of Each Species:")
display(df['species'].value_counts())
```

```python
# 2. Univariate Analysis

# Histograms
df.hist(figsize=(10, 8), bins=15)
plt.suptitle("Histogram of Iris Features")
plt.show()
```

```python
# Boxplots by species
df.boxplot(column="sepal_length", by="species", figsize=(8, 6))
plt.title("Sepal Length Distribution by Species")
plt.suptitle("")
plt.show()
```

```python
# 3. Bivariate Analysis

# Scatter plot sepal vs petal length
colors = {"setosa": "red", "versicolor": "green", "virginica": "blue"}
df.plot(kind="scatter", x="sepal_length", y="petal_length",
        c=df["species"].map(colors), figsize=(8, 6))
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
# Scatter matrix (pairwise relationships)
pd.plotting.scatter_matrix(df.iloc[:, :-1], figsize=(12, 10), diagonal="hist",
                           c=df["species"].map(colors))
plt.suptitle("Pairwise Relationships Between Features")
plt.show()
```

```python
# 4. Grouped Statistics

print("\nðŸ”¹ Mean values grouped by species:")
display(df.groupby("species").mean())

print("\nðŸ”¹ Min values grouped by species:")
display(df.groupby("species").min())

print("\nðŸ”¹ Max values grouped by species:")
display(df.groupby("species").max())
```
